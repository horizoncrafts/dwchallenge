```python
import gzip
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim.models import Word2Vec, LdaMulticore

from gensim.corpora import Dictionary

import numpy as np
import pyLDAvis
from pyLDAvis import gensim
```

```python
def preproces_sentence(sentence):
    return [word for word in simple_preprocess(sentence) if word not in STOPWORDS]

def read_sentences(filename):
    with open(filename, 'rb') as f:
        for line in f:
            yield preproces_sentence(line)
        
%time sentences = list(read_sentences('/data/notebooks/challenge04/day03/data/reviews_data.txt'))
```

```python
len(sentences)
```

```python
%time model = Word2Vec(sentences, size=100, window=5, min_count=2)
```

```python
model.wv.most_similar(positive=['downtown'])
```

## Topic modeling - LDA

```python
sentences_light = np.random.permutation(sentences)
```

```python
sentences_light = sentences_light[:2000]
```

```python
dictionary = Dictionary(sentences_light)
```

```python
list(dictionary.items())
```

```python
bow_corpus = [ dictionary.doc2bow(sentence) for sentence in sentences_light]
```

```python
bow_corpus[0]
```

```python
%time lda_model = LdaMulticore(bow_corpus, id2word=dictionary, num_topics=10, passes=20, workers=4)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print( f'Topic {idx} \nWords {topic}')
```

## Wizualizacja

```python
lda_vis = pyLDAvis.gensim.prepare(lda_model, bow_corpus, dictionary)
pyLDAvis.display(lda_vis)
```
